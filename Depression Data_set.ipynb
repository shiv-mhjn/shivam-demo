# Insight 11 Plotting a countplot on Smoking status and Effect on Mental Illness

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(data=df, x='Smoking Status', hue="History of Mental Illness")
plt.title('Smoking Status by History of Mental Illness')
plt.xlabel('Smoking Status')
plt.ylabel('Count')
plt.legend(title='History of Mental Illness')
plt.show()
```
